```python
[4]
# There is a very problematic section that doesn't react to clicking via Selenium
# this function is a workaround that uses TAB-ENTER combination to locate the button from that section and click it
from selenium.webdriver.common.action_chains import ActionChains

actions = ActionChains(driver)

def locate_and_click(tabs):
    #actions.reset_actions()
    for _ in range(tabs):                       # Press Tab {tabs} times
        actions.send_keys(Keys.TAB)
        time.sleep(0.05)
        actions.perform()
    actions.send_keys(Keys.ENTER)       # Press Enter
    actions.perform()                   # Perform the actions
```

```python
[6]
# Oh, well, we'll process what we've got now and create a list of the companies we do have
# (As well as create a function to process excel files and create the stock dataframe)

import os
import re




combined_data = pd.DataFrame()

# Function to extract ticker and company code from the cell value
def extract_ticker_and_code(cell_value):
    ticker_pattern = re.compile(r'^[A-Z0-9]+:[A-Z0-9]+')
    code_pattern = re.compile(r'MI Key: (\d+)')
    
    ticker_match = ticker_pattern.search(cell_value)
    code_match = code_pattern.search(cell_value)
    
    if ticker_match and code_match:
        ticker = ticker_match.group(0)
        company_code = code_match.group(1)
        return ticker, company_code
    else:
        return None, None

# Iterate over all Excel files in the folder
def go_through_files(folder_path = r'C:\Users\panov\Downloads\Excels for companies')

    for file_name in os.listdir(folder_path):
        if file_name.endswith('.xlsx') or file_name.endswith('.xls'):
            file_path = os.path.join(folder_path, file_name)
            
            # Read the relevant data range from the Excel file, skipping metadata rows
            df = pd.read_excel(file_path, sheet_name=0, skiprows=43, nrows=784, usecols="A:D", header=None)
            
            # Manually set the column names to ensure consistency
            df.columns = ['Date', 'Close', 'Volume', 'Industry Group']
            
            # Format the date column and set it as the index
            df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
            df.set_index('Date', inplace=True)
            
            # Extract the ticker and company code from cell A4
            ticker_and_code = pd.read_excel(file_path, sheet_name=0, usecols="A", nrows=4).iloc[2, 0]
            print(ticker_and_code)
            ticker, company_code = extract_ticker_and_code(ticker_and_code)
            
            # Add the extracted ticker and company code as new columns
            df['Ticker'] = ticker
            df['Company code'] = company_code
            
            # Append the data to the combined DataFrame
            combined_data = pd.concat([combined_data, df])

    return combined_data


combined_data.reset_index(inplace=True)
```
