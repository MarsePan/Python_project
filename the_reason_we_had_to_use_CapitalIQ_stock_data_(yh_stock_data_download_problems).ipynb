```python
# Load the article data
import pandas as pd 

articles_1 = pd.read_csv("company_articles.csv")
articles_2 = pd.read_csv("company_articles_updated.csv")

articles = pd.concat([articles_1, articles_2], ignore_index=True)
articles
```

```python
# Get the unique company codes
unique_company_codes = pd.DataFrame(articles['company_code'].unique(), columns=['company_code'])
unique_company_codes
```

```python
# Get the unique company codes as ticks
companies_50_perc_df = pd.read_excel("companies_50_perc.xlsx", engine='openpyxl')
companies_df = pd.read_excel("companies.xlsx", engine='openpyxl')
companies_list = pd.concat([companies_df, companies_50_perc_df], ignore_index=True)
companies_list

companies_list = companies_list[companies_list['Entity ID'].isin(unique_company_codes['company_code'])].sort_values(by='Entity ID')
tickers = companies_list["Ticker"].unique()
tickers
```

```python
# Some problems with MarketCap values - will prob need to reload later again
duplicates = companies_list[companies_list.duplicated('Entity ID', keep=False)]
duplicates
```

```python
# We'll convert dates before proceeding
articles['publication_date'] = pd.to_datetime(articles['publication_date'], format='%A, %B %d, %Y %I:%M %p ET')
articles
```

```python
# Get stock data
import yfinance as yf
from datetime import timedelta

start_date = min(articles['publication_date']) - timedelta(days=60)
end_date = max(articles['publication_date']) + timedelta(days=60)

stocks = pd.DataFrame(columns=['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume', 'Ticker'])
not_found_tickers = []

for i in tickers:
    try:
        stock_current = yf.download(i, start=start_date, end=end_date)
        if stock_current.empty:
            not_found_tickers.append(i)
        else:
            stock_current["Ticker"] = i
            stocks = pd.concat([stocks, stock_current], ignore_index=True)
    except Exception as e:
        not_found_tickers.append(i)
        
stocks
```
